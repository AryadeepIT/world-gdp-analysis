## Preprocessing

```python
import pandas as pd   
import numpy as np   
import matplotlib.pyplot as plt
import plotly.express as px

# !pip install pycountry
import pycountry # country locales for filtering non-countries


# !pip install pandas-bokeh 
import pandas_bokeh
from bokeh.models.widgets import DataTable, TableColumn
from bokeh.models import ColumnDataSource

pandas_bokeh.output_notebook()

%matplotlib inline
```

## Load Dataset

```python
df = pd.read_csv('gdp-1960-2021.csv')
fdf = pd.read_csv('gdp-1960-2021.csv').fillna(0)
display(df.head(5))
display(df.describe())
```

## Data Analysis of Top 10 Countries according to 2021

Top 10 GDPs of 2021 are for non-countries, so filtering to only include countries in descending order using the library pycountry.

```python
print('\n ⨳ Top 10 List Based on Non-Country : ')
display(df[['Country Name', 'Country Code', '2021']].sort_values('2021', ascending=False).head(10))

countries_code = [x.alpha_3 for x in pycountry.countries]
result = df['Country Code'].isin(countries_code)
list_country = []
for i in range(len(result)):
    if(result[i] == False):
        list_country.append(i)

print(f'\n ⨳ Number of Countries : {len(list_country)}')
df=df.drop(list_country, axis=0)

print('\n ⨳ New top 10 Countries of 2021 :')
df[['Country Name', 'Country Code', '2021']].sort_values('2021', ascending=False).head(10)
```

## Removing Not Useful Columns

```python
print('\n ⨳ Dataset View : ')
display(df.head(3))

# display(df[['Indicator Code','Indicator Name', '2020']].groupby('Indicator Name').count())

print("\n ⨳ 'Indicator Code', 'Indicator Name' don't contain anything useful so I'm removing those.")
print("\n ⨳ 'Country Code' is not useful as 'Country Name' Column serves the same purpose. so I'm removing these.")

del df['Indicator Name']
del df['Indicator Code']
del df['Country Code']

print('\n Removed 3 Columns ✔')
```

### Checking for Duplicate Values in Country Name Column

```python
print('\n ⨳ No Duplicate Countries Present. :D')
df[df['Country Name'].duplicated()]
```

## Getting Data Ready

⨳ Working with data that has date as the index using pd.melt() and pd.pivot()

```python
melt_df = pd.melt(df, id_vars=["Country Name"], var_name="Year", value_name="GDP")
display(melt_df.head())

pivot_df = melt_df.pivot(index='Year', columns='Country Name', values='GDP')
display(pivot_df.head())
```

 ⨳ A list of the top and bottom 10 countries based on 2021 GDP for future use.

```python
top_10_countries = pivot_df.loc['2021':].T.dropna().sort_values(by='2021', ascending=False).head(10).index.tolist()
bottom_10_countries = pivot_df.loc['2021':].T.dropna().sort_values(by='2021').head(10).index.tolist()
print("Top 10 Countries according to 2021 GDP : ",top_10_countries)
print()
print("Bottom 10 Countries according to 2021 GDP : ",bottom_10_countries)
```

### Converting Year to datetime : 

```python
pivot_df = pivot_df.reset_index()
pivot_df['Year'] = pd.to_datetime(pivot_df['Year'])
pivot_df = pivot_df.set_index('Year')
pivot_df.tail()
```

```python
dates_2011_2021 = pivot_df.index.to_series().between('2011-01-01', '2021-01-01')
#dates_1960_2021 = df.index.to_series().between('1960-01-01', '2021-01-01')

past_11_years = pivot_df[dates_2011_2021]

past_11_years[top_10_countries].head(11)
```

### Analyze Percentage Growth

```python
pct_change = pivot_df.pct_change()
dates_1970_2021 = pct_change.index.to_series().between('1970-01-01', '2021-01-01')
pct_change = pct_change[dates_1970_2021].multiply(100)
display(pct_change.tail())

cumulative_pct_change = pct_change.dropna(axis='columns').cumsum().tail(1).reset_index().drop(['Year'], axis=1).T.sort_values(by=0, ascending=False).rename(columns = {0:"Cumulative % Change in GDP"})
cumulative_pct_change
```

```python
dates_1970_2021 = pct_change.index.to_series().between('1970-01-01', '2021-01-01')
temp = pct_change[dates_1970_2021].dropna(axis='columns')

top_5 = cumulative_pct_change.index.tolist()[:5]
top_5.append('India')

temp = temp[top_5]

p_line = temp.plot_bokeh.line(
    title="% Change in GDP - Top 5 Countries + India",
    rangetool=True
)
```

```python
dates_1970_2021 = pct_change.index.to_series().between('1970-01-01', '2021-01-01')
temp = pct_change[dates_1970_2021].dropna(axis='columns')

bottom_5 = cumulative_pct_change.index.tolist()[-5:]
bottom_5.append('India')

temp = temp[bottom_5]

p_line = temp.plot_bokeh.line(
    title="% Change in GDP - Bottom 5 Countries + India",
    rangetool=True
)
```

```python
pct_change_world = pct_change.mean(axis=1)
pct_change_world = pd.DataFrame(pct_change_world, columns=['Average GDP Change'])
display(pct_change_world.head())

p_line = pct_change_world.plot_bokeh.line(
    title="Average % Change in GDP for All Countries",
    rangetool=True
)
```
